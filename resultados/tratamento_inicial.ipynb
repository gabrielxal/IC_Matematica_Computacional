```python
import pandas as pd
import numpy as np
```

# MMQ junção dos dados

```python
dfMMQ = pd.read_excel('comportamentoMMQ_preliminares.xlsx')
```

## MP tratamento dos dados

```python
dfMMQ = dfMMQ.drop(columns=dfMMQ.columns[dfMMQ.columns.str.contains('^Unnamed')])
```

```python
dfMMQ['Matriz'] = dfMMQ['Matriz'].str.replace('.mtx', '')
dfMMQ['Matriz'] = dfMMQ['Matriz'].apply(lambda x: x.split('/')[-1])
```

```python
dfMMQ
```

```python
dfMMQ2 = dfMMQ[dfMMQ['r2'] != 'erro']
```

```python
df2MMQ = dfMMQ2.query('r2 > 0.45')
df2MMQ = df2MMQ.query('Metodo == "MP" ')
df2MMQ
```

```python
valores_unicos = df2MMQ['Matriz'].unique()
```

```python
valores_unicos.size
```

```python
valores_unicos
```

```python
#salvar
df2MMQ.to_excel('tabelas_final/resultado_MMQ_final.xlsx')
```

# MP junção dos dados

```python
dfMP = pd.read_excel('resultados_preliminares.xlsx')
```

## MP tratamento dos dados

```python
dfMP = dfMP.drop(columns=dfMP.columns[dfMP.columns.str.contains('^Unnamed')])
```

```python
dfMP['Matriz'] = dfMP['Matriz'].str.replace('.mtx', '')
dfMP['Matriz'] = dfMP['Matriz'].apply(lambda x: x.split('/')[-1])
```

```python
dfMP = dfMP.query('Método == "MP"')
```

```python
dfMP
```

## Olhar Matrizes com alto r2

```python
linhas_filtradas = dfMP[dfMP['Matriz'].isin(valores_unicos)]
```

```python
linhas_filtradas
```

```python
#salvar
dfMP.to_excel('tabelas_final/resultado_MP_final.xlsx')
linhas_filtradas.to_excel('tabelas_final/resultado_MP_MMQ.xlsx')
```
